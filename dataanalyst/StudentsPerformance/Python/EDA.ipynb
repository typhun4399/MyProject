```python
group = ["gender", "race/ethnicity", "parental level of education", "lunch", "test preparation course"]
score = ["math score", "reading score", "writing score"]

for g in group:
    print(f"{g} : {df[g].unique()}")
```

```python
df[score].describe()
```

```python
for g in group:
    for col in score:
        print(f"Average {col} grouped by {g}")
        print(df.groupby(g)[col].mean().sort_values(ascending = False))
    print("-" * 50)
```
